```python
import pandas as pd
```

```python
df = pd.read_csv("LB_TA.txt", delimiter = r"\s+")
df
```

menampilkan data dalam bentuk txt

```python
df = df.drop_duplicates()
df
```

menghapus data yang duplikat

```python
df = df.dropna(subset=['at'])
df
```

menghapus data yang memiliki nilai kosong pada kolom at

```python
df = df.groupby('tic')['at'].mean().reset_index()
df
```

Menghitung rata-rata nilai di kolom at untuk nilai yang sama pada kolom tic

```python
df = df.rename(columns={'at': 'avg_at'})
df
```

mengubah nama kolom at menjadi avg_at

```python
df.to_csv("Hasil Data.csv", index = False)
df
```
